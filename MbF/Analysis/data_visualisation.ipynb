### Compare two targets

```python
main_file = 'mbf'

name_1 = 'SLSN-I+host'
features_1 = pd.read_parquet(f'../data/features/{main_file}/{name_1}_features.parquet')
features_1['target'] = name_1

name_2 = 'TDE'
features_2 = pd.read_parquet(f'../data/features/{main_file}/{name_2}_features.parquet')
features_2['target'] = name_2

features = pd.concat([features_1, features_2])
```

```python
param_names = list(features.keys())

for i in features.keys():
    plt.figure()
    plt.title(i)
    
    zoom = zoom_in(features, 0.99, param_names)
    zoom_features = features[zoom]
    
    nbins = 75
    
    for target in np.unique(features['target']):
        plt.hist(zoom_features[zoom_features['target']==target][i], bins = nbins, alpha=0.5, density=True, label=target);
    
        plt.legend()
features.keys()
```
